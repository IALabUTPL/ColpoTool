# üî¨ Clasificaci√≥n de Lesiones Cervicales con Detectron2

## üìÑ Descripci√≥n

Este cuaderno tiene como objetivo desarrollar un modelo de aprendizaje profundo utilizando la arquitectura Mask R-CNN con Detectron2, orientado a la **clasificaci√≥n y detecci√≥n de lesiones cervicales** en im√°genes biom√©dicas. Para ello, se parte de anotaciones previamente generadas en formato **COCO JSON** o **VGG VIA**. Se implementan procesos de unificaci√≥n, conversi√≥n, registro de datasets, visualizaci√≥n, entrenamiento del modelo y an√°lisis de resultados.

Detectron2 ofrece una soluci√≥n robusta y altamente optimizada para tareas de detecci√≥n y segmentaci√≥n de instancias, y su uso en este contexto busca automatizar y mejorar la precisi√≥n del diagn√≥stico colposc√≥pico.

---

## üéØ Objetivos

- Unificar anotaciones de diferentes fuentes (COCO, VGG VIA) en un √∫nico archivo est√°ndar.
- Registrar datasets personalizados en Detectron2 para su uso en entrenamiento y validaci√≥n.
- Entrenar un modelo Mask R-CNN desde `model_zoo` para clasificaci√≥n y segmentaci√≥n de lesiones.
- Evaluar el rendimiento del modelo en t√©rminos de precisi√≥n y detecci√≥n por clase.
- Exportar segmentaciones, predicciones y m√©tricas en formatos visuales y num√©ricos.

---


## üìÅ Dataset

- **Formato original**: COCO JSON, VGG VIA JSON.
- **Etiquetas esperadas**: `demy`, `isch`, entre otras posibles relacionadas a tipos de lesi√≥n.
- **Origen**: Im√°genes biom√©dicas de cervix provenientes de estudios cl√≠nicos (Intel dataset, KAIME, etc.).

```python
import json
from pathlib import Path
from collections import Counter
import matplotlib.pyplot as plt

# üìÅ Rutas reales
BASE_DIR = Path.cwd()
INTEL_DIR = BASE_DIR / "ColpoToolLab" / "intel"
JSON_DIR = BASE_DIR / "ColpoToolLab" / "detectron_data"

print(f"üìÇ INTEL_DIR: {INTEL_DIR}")

# üßæ JSONs a usar
json_paths = [
    JSON_DIR / "colpo_2_coco.json",
    JSON_DIR / "coco_labels_32_aw_m_p_av_2025-04-10-07-28-25.json"
]

# üìÇ Paso 1: Listar im√°genes reales en carpeta Intel
imagenes_intel = {img.name for img in INTEL_DIR.glob("*.jpg")}
print(f"üñºÔ∏è Total im√°genes .jpg en Intel: {len(imagenes_intel)}")
print(f"üîç Ejemplos: {list(imagenes_intel)[:5]}")

# üîÑ Paso 2: Cargar JSONs y filtrar im√°genes v√°lidas
imagenes_validas = {}
anotaciones_validas = []

for json_file in json_paths:
    print(f"\nüì¶ Procesando: {json_file.name}")
    if not json_file.exists():
        print(f"‚ùå Archivo no encontrado: {json_file}")
        continue

    with open(json_file, 'r') as f:
        data = json.load(f)

    images_json = data.get("images", [])
    annotations_json = data.get("annotations", [])

    print(f"üì∑ Im√°genes en JSON: {len(images_json)} | üß© Anotaciones: {len(annotations_json)}")

    # Validar im√°genes que existen f√≠sicamente
    for img in images_json:
        nombre = Path(img["file_name"]).name
        ruta_imagen = INTEL_DIR / nombre

        if ruta_imagen.exists():
            imagenes_validas[img["id"]] = nombre
        else:
            print(f"‚ö†Ô∏è Imagen ausente en disco: {nombre}")

    # Validar anotaciones cuya imagen existe
    for ann in annotations_json:
        if ann["image_id"] in imagenes_validas:
            nombre_img = imagenes_validas[ann["image_id"]]
            ruta_img = INTEL_DIR / nombre_img

            if ruta_img.exists():  # Reconfirmaci√≥n (doble filtro)
                segmentacion = ann.get("segmentation", [])
                if isinstance(segmentacion, list) and len(segmentacion) > 0 and len(segmentacion[0]) >= 6:
                    anotaciones_validas.append({
                        "nombre": nombre_img,
                        "categoria_id": ann["category_id"],
                        "segmentacion": segmentacion
                    })
                else:
                    print(f"‚ö†Ô∏è Anotaci√≥n inv√°lida (segmentaci√≥n vac√≠a o corta): {nombre_img}")

# üìä Paso 3: Conteo por categor√≠a
conteo = Counter(ann["categoria_id"] for ann in anotaciones_validas)

# üßæ Resultados
print(f"\n‚úÖ Total im√°genes v√°lidas (con al menos una anotaci√≥n v√°lida): {len(set(ann['nombre'] for ann in anotaciones_validas))}")
print(f"‚úÖ Total anotaciones v√°lidas: {len(anotaciones_validas)}")

# üß™ Paso adicional: dividir en train y val
from sklearn.model_selection import train_test_split

# Obtener nombres √∫nicos de im√°genes
imagenes_unicas = sorted(list(set(ann["nombre"] for ann in anotaciones_validas)))

# Dividir en 80% train, 20% val
train_imgs, val_imgs = train_test_split(imagenes_unicas, test_size=0.2, random_state=42)

# Asignar a cada anotaci√≥n su conjunto
anotaciones_train = [a for a in anotaciones_validas if a["nombre"] in train_imgs]
anotaciones_val = [a for a in anotaciones_validas if a["nombre"] in val_imgs]


print("\nüìä Conteo por clase:")
for clase, cantidad in conteo.items():
    print(f" - Clase {clase}: {cantidad}")

# üß™ Muestra
print("\nüß™ Muestra de 5 anotaciones:")
for ann in anotaciones_validas[:5]:
    print(f" - Imagen: {ann['nombre']}, Clase: {ann['categoria_id']}, Segmento: {str(ann['segmentacion'])[:40]}...")

# üìà Visualizaci√≥n
if not conteo:
    print("‚ö†Ô∏è No hay datos para graficar.")
else:
    clases = list(conteo.keys())
    cantidades = list(conteo.values())

    plt.figure(figsize=(8, 5))
    plt.bar(clases, cantidades, color='cornflowerblue', edgecolor='black')
    plt.xlabel("ID de Clase")
    plt.ylabel("Cantidad de Anotaciones")
    plt.title("Distribuci√≥n de clases en las anotaciones v√°lidas")
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
import cv2
from detectron2.data.datasets import register_coco_instances
from detectron2.data import MetadataCatalog, DatasetCatalog
from PIL import Image

# ‚úÖ Inicializaci√≥n
final_images = []
final_annotations = []
img_id_map = {}
img_id_counter = 1
ann_id_counter = 1

for ann in anotaciones_validas:
    nombre = ann["nombre"]
    segmentacion = ann["segmentacion"]

    # ‚ùå Omitir anotaciones vac√≠as o inv√°lidas
    if not isinstance(segmentacion, list) or len(segmentacion) == 0 or len(segmentacion[0]) < 6:
        print(f"‚ö†Ô∏è Anotaci√≥n inv√°lida omitida: {nombre}")
        continue

    # üñºÔ∏è Procesar imagen si no est√° registrada
    if nombre not in img_id_map:
        img_path = str(INTEL_DIR / nombre)

        imagen = cv2.imread(img_path)
        if imagen is None:
            print(f"‚ö†Ô∏è Imagen no encontrada o corrupta: {nombre}")
            continue
        height, width = imagen.shape[:2]  # OpenCV da (alto, ancho)


        final_images.append({
            "id": img_id_counter,
            "file_name": nombre,
            "height": height,
            "width": width
        })
        img_id_map[nombre] = img_id_counter
        img_id_counter += 1

    # üì¶ Calcular bbox y √°rea
    puntos = segmentacion[0]
    x_coords = puntos[0::2]
    y_coords = puntos[1::2]
    if not x_coords or not y_coords:
        print(f"‚ö†Ô∏è Coordenadas vac√≠as, omitido: {nombre}")
        continue
    x_min, x_max = min(x_coords), max(x_coords)
    y_min, y_max = min(y_coords), max(y_coords)
    bbox = [x_min, y_min, x_max - x_min, y_max - y_min]
    area = bbox[2] * bbox[3]

    final_annotations.append({
        "id": ann_id_counter,
        "image_id": img_id_map[nombre],
        "category_id": ann["categoria_id"],
        "segmentation": segmentacion,
        "iscrowd": 0,
        "bbox": bbox,
        "area": area
    })
    ann_id_counter += 1

# ‚úÖ Categor√≠as
final_categories = [
    {"id": 1, "name": "aw"},
    {"id": 2, "name": "m"},
    {"id": 3, "name": "p"},
    {"id": 4, "name": "av"},
    {"id": 5, "name": "y"}
]

from PIL import Image

def get_colpo_dicts_train():
    return construir_dataset(anotaciones_train)

def get_colpo_dicts_val():
    return construir_dataset(anotaciones_val)

def construir_dataset(anotaciones_subset):
    dataset_dicts = []
    nombres_procesados = set()
    img_id_counter = 0  # Contador incremental para IDs consistentes

    for ann in anotaciones_subset:
        nombre = ann["nombre"]
        img_path = INTEL_DIR / nombre
        if not img_path.exists() or nombre in nombres_procesados:
            continue

        imagen = cv2.imread(str(img_path))
        if imagen is None:
            continue

        height, width = imagen.shape[:2]
        record = {
            "file_name": str(img_path),
            "image_id": img_id_counter,  # ID consistente, no variable como hash()
            "height": height,
            "width": width,
            "annotations": []
        }

        for sub_ann in [a for a in anotaciones_subset if a["nombre"] == nombre]:
            puntos = sub_ann["segmentacion"][0]

            # Validar puntos fuera de rango
            if max(puntos) > 1.5 * max(width, height):
                print(f"‚ö†Ô∏è Coordenadas fuera de escala en {nombre} ‚Üí max punto: {max(puntos)} | Tama√±o: {width}x{height}")

            x_coords = puntos[0::2]
            y_coords = puntos[1::2]
            bbox = [min(x_coords), min(y_coords), max(x_coords) - min(x_coords), max(y_coords) - min(y_coords)]

            record["annotations"].append({
                "bbox": bbox,
                "bbox_mode": 0,  # BoxMode.XYWH_ABS
                "segmentation": [puntos],
                "category_id": sub_ann["categoria_id"] - 1,
                "iscrowd": 0
            })

        dataset_dicts.append(record)
        nombres_procesados.add(nombre)
        img_id_counter += 1  # Aumentar el contador por cada imagen registrada

    return dataset_dicts


print("üîé Verificando segmentaciones mal formadas...")

errores_seg = 0
for d in get_colpo_dicts_train():
    for ann in d["annotations"]:
        if len(ann["segmentation"][0]) < 6:
            print(f"‚ùå Segmentaci√≥n inv√°lida en {d['file_name']} con categor√≠a {ann['category_id']}")
            errores_seg += 1

print(f"‚úÖ Verificaci√≥n finalizada. Total de segmentaciones inv√°lidas: {errores_seg}")

print("üîé Verificando IDs de categor√≠a...")
ids = [cat["id"] for cat in final_categories]
if sorted(ids) != list(range(len(final_categories))):
    print("‚ùå Los ID de categor√≠as no est√°n en secuencia desde 0. Esto puede causar errores.")
else:
    print("‚úÖ Los ID de categor√≠as est√°n correctamente definidos.")

from detectron2.data import DatasetCatalog, MetadataCatalog

# üîÅ Eliminar si ya est√°n registrados
for dname in ["colpo_train", "colpo_val"]:
    if dname in DatasetCatalog.list():
        DatasetCatalog.remove(dname)

# ‚úÖ Registrar nuevamente
DatasetCatalog.register("colpo_train", get_colpo_dicts_train)
DatasetCatalog.register("colpo_val", get_colpo_dicts_val)

MetadataCatalog.get("colpo_train").set(thing_classes=[c["name"] for c in final_categories])
MetadataCatalog.get("colpo_val").set(thing_classes=[c["name"] for c in final_categories])

print("‚úÖ Datasets registrados: colpo_train y colpo_val.")


MetadataCatalog.get("colpo_train").set(thing_classes=[cat["name"] for cat in final_categories])
print("‚úÖ Dataset en memoria registrado como 'colpo_train'.")

```

```python
import random
import cv2
import matplotlib.pyplot as plt
from detectron2.utils.visualizer import Visualizer

def visualizar_muestras(dataset_dicts, metadata, num_muestras=5, titulo="Muestras"):
    plt.figure(figsize=(15, 5 * num_muestras))
    count = 0
    for d in random.sample(dataset_dicts, min(num_muestras, len(dataset_dicts))):
        img_path = d["file_name"]
        img = cv2.imread(img_path)

        if img is None:
            print(f"‚ùå Imagen no encontrada: {img_path}")
            continue

        visualizer = Visualizer(img[:, :, ::-1], metadata=metadata, scale=0.5)
        vis = visualizer.draw_dataset_dict(d)
        plt.subplot(num_muestras, 1, count + 1)
        plt.imshow(vis.get_image()[:, :, ::-1])
        plt.title(f"{titulo} #{count+1}")
        plt.axis('off')
        count += 1

    if count == 0:
        print("‚ö†Ô∏è No se pudieron visualizar muestras (todas fallaron).")
    else:
        plt.tight_layout()
        plt.show()
from pathlib import Path


print("üìÅ INTEL_DIR:", INTEL_DIR)
print("üñºÔ∏è Total im√°genes:", len(list(INTEL_DIR.glob("*.jpg"))))

# Ejecutar visualizaci√≥n
visualizar_muestras(DatasetCatalog.get("colpo_train"), MetadataCatalog.get("colpo_train"), num_muestras=5, titulo="Visualizaci√≥n - Dataset Completo")
```

```python
import os
import torch
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2 import model_zoo

# üìÅ Ruta de salida para modelos entrenados
OUTPUT_DIR = "output/colpo_model"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ‚öôÔ∏è Configurar modelo
cfg = get_cfg()
cfg.OUTPUT_DIR = OUTPUT_DIR

# üß† Usar arquitectura base del modelo
cfg.merge_from_file(model_zoo.get_config_file(
    "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))

# üì¶ Cargar solo pesos del backbone (sin cabeza entrenada)
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url(
    "COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")

# üìä Datasets personalizados
cfg.DATASETS.TRAIN = ("colpo_train",)
cfg.DATASETS.TEST = ("colpo_val",)

# üéØ Clases personalizadas
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 5
cfg.MODEL.ROI_HEADS.NAME = "StandardROIHeads"  # üëà Evita conflictos con cabeza de COCO

# üß™ Umbral de puntuaci√≥n para predicciones
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.05

# üöÄ Dispositivo: GPU
cfg.MODEL.DEVICE = "cuda:0"

# üßÆ Hiperpar√°metros del optimizador
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 3000  # üëà Aumentado para que realmente aprenda
cfg.SOLVER.STEPS = []  # Sin reducci√≥n de LR
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 256

# üß™ Evaluar cada cierto n√∫mero de iteraciones
cfg.TEST.EVAL_PERIOD = 500

# ‚úÖ Entrenar desde cero (no resume)
trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```

```python
import os
import json
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
from pathlib import Path

import torch
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor, DefaultTrainer
from detectron2 import model_zoo
from detectron2.data import DatasetCatalog, MetadataCatalog, build_detection_test_loader
from detectron2.evaluation import COCOEvaluator, inference_on_dataset
from detectron2.utils.visualizer import Visualizer, ColorMode
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

# ==== CONFIGURACI√ìN CON CHECKPOINT ENTRENADO ====
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.OUTPUT_DIR = "output/colpo_model"
cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
cfg.DATASETS.TEST = ("colpo_val",)
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 5
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.05
cfg.MODEL.DEVICE = "cuda:0"

# Registrar clases
MetadataCatalog.get("colpo_val").thing_classes = ["aw", "m", "p", "av", "y"]
predictor = DefaultPredictor(cfg)

# ==== EVALUACI√ìN ====
evaluator = COCOEvaluator("colpo_val", output_dir=cfg.OUTPUT_DIR)
val_loader = build_detection_test_loader(cfg, "colpo_val")
results = inference_on_dataset(predictor.model, val_loader, evaluator)
print("\nüìä Resultados de evaluaci√≥n:")
print(results)

# ==== AP por clase (manual) ====
COCO_JSON_PATH = Path("output/colpo_model/colpo_val_coco_format.json")
coco_gt = COCO(str(COCO_JSON_PATH))
coco_dt = coco_gt.loadRes("output/colpo_model/coco_instances_results.json")
coco_eval = COCOeval(coco_gt, coco_dt, iouType="segm")
coco_eval.evaluate()
coco_eval.accumulate()
coco_eval.summarize()

cats = coco_gt.loadCats(coco_gt.getCatIds())
class_names = [c["name"] for c in cats]
ap_results = []
for idx, name in enumerate(class_names):
    precision = coco_eval.eval["precision"][:, :, idx, 0, 0]
    precision = precision[precision > -1]
    ap = np.mean(precision) if precision.size else float("nan")
    ap_results.append((name, ap))

# ==== HEATMAP ====
df_ap = pd.DataFrame(ap_results, columns=["Clase", "AP"]).dropna()
sns.heatmap(df_ap.set_index("Clase").T, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("üìä AP por Clase (calculado manualmente)")
plt.show()

# ==== VISUALIZACI√ìN ====
dataset_val = DatasetCatalog.get("colpo_val")
for d in random.sample(dataset_val, 5):
    im = cv2.imread(d["file_name"])
    outputs = predictor(im)
    v = Visualizer(im[:, :, ::-1],
                   MetadataCatalog.get("colpo_val"),
                   scale=0.5,
                   instance_mode=ColorMode.IMAGE_BW)
    out = v.draw_instance_predictions(outputs["instances"].to("cpu"))
    cv2.imshow("Predicci\u00f3n", out.get_image()[:, :, ::-1])
    cv2.waitKey(0)
cv2.destroyAllWindows()

# ==== REVISI√ìN DE CONFIGURACI√ìN ====
def revisar_configuracion(cfg):
    print("\nüß† REVISI√ìN DE CONFIGURACI√ìN")
    print("üì¶ Pesos cargados desde:", cfg.MODEL.WEIGHTS)
    print("üìä N√∫mero de clases:", cfg.MODEL.ROI_HEADS.NUM_CLASSES)
    print("üìÅ Dataset VAL:", cfg.DATASETS.TEST)
    print("üéØ Umbral de puntuaci√≥n:", cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST)
    print("üìé Categor√≠as definidas:", MetadataCatalog.get(cfg.DATASETS.TEST[0]).thing_classes)
    if cfg.MODEL.ROI_HEADS.NUM_CLASSES != len(MetadataCatalog.get(cfg.DATASETS.TEST[0]).thing_classes):
        print("‚ùå MISMATCH entre NUM_CLASSES y categor√≠as registradas. Verifica el registro de metadata.")
    else:
        print("‚úÖ N√∫mero de clases y categor√≠as est√°n alineados.")

revisar_configuracion(cfg)
```

```python
import json
from pathlib import Path
from detectron2.data import DatasetCatalog, MetadataCatalog

# Rutas
COCO_JSON_PATH = Path("output/colpo_model/colpo_val_coco_format.json")
VAL_IMG_DIR = Path("/tf/workspace/ColpoToolLab/intel")

# üîÅ Construir un mapa consistente entre nombre e ID
# ‚ö†Ô∏è Reutiliza esto tambi√©n al construir el JSON COCO en la parte de anotaciones
print("\nüîß Generando img_id_map a partir de nombres √∫nicos en el dataset 'colpo_val'...")
dataset_val = DatasetCatalog.get("colpo_val")

nombres_imagenes = sorted(list(set([Path(d["file_name"]).name for d in dataset_val])))
img_id_map = {nombre: i for i, nombre in enumerate(nombres_imagenes)}

# Mostrar el mapa generado
print("üó∫Ô∏è Muestra del img_id_map (primeros 5):")
for k in list(img_id_map.keys())[:5]:
    print(f"  {k} ‚Üí ID {img_id_map[k]}")

# ‚úÖ Verificar los ID actuales en dataset_val
print(f"\nüîç Total de elementos en 'colpo_val': {len(dataset_val)}")
print("\nüì∑ Ejemplos del dataset 'colpo_val':")
for i, d in enumerate(dataset_val[:3]):
    nombre = Path(d["file_name"]).name
    id_esperado = img_id_map.get(nombre, "‚ùå No encontrado")
    print(f"  [{i}] ID actual: {d['image_id']} | Esperado: {id_esperado} | Archivo: {nombre}")
    for ann in d["annotations"]:
        print(f"    - Clase: {ann['category_id']} | Segmentaci√≥n: {str(ann['segmentation'])[:30]}...")

# ‚úÖ Paso 2: Verificar unicidad de los IDs en el dataset
ids_dataset = [d["image_id"] for d in dataset_val]
if len(set(ids_dataset)) == len(ids_dataset):
    print("\n‚úÖ Todos los 'image_id' en 'colpo_val' son √∫nicos.")
else:
    print("\n‚ùå Hay duplicados en los 'image_id' del dataset.")

# ‚úÖ Paso 3: Cargar el JSON exportado (si existe)
if COCO_JSON_PATH.exists():
    print(f"\nüìÇ Archivo encontrado: {COCO_JSON_PATH.name}")
    with open(COCO_JSON_PATH, 'r') as f:
        coco_data = json.load(f)

    ids_json = [img["id"] for img in coco_data.get("images", [])]
    print(f"üìä Total im√°genes en COCO JSON: {len(ids_json)}")

    # ‚úÖ Paso 4: Comparar los IDs
    print("\nüîç Comparando 'image_id' entre dataset y JSON exportado...")
    ids_dataset_set = set(ids_dataset)
    ids_json_set = set(ids_json)

    missing_in_json = ids_dataset_set - ids_json_set
    missing_in_dataset = ids_json_set - ids_dataset_set

    if missing_in_json:
        print(f"‚ö†Ô∏è IDs en dataset que NO est√°n en el JSON: {sorted(missing_in_json)}")
    else:
        print("‚úÖ Todos los IDs del dataset est√°n en el JSON.")

    if missing_in_dataset:
        print(f"‚ö†Ô∏è IDs en JSON que NO est√°n en el dataset: {sorted(missing_in_dataset)}")
    else:
        print("‚úÖ Todos los IDs del JSON est√°n en el dataset.")
else:
    print(f"\n‚ùå No se encontr√≥ el archivo: {COCO_JSON_PATH}")
    print("   ‚Üí Aseg√∫rate de que se haya generado correctamente al evaluar.")

```

```python
# ‚úÖ Si los IDs no coinciden, regenerar el archivo COCO JSON con IDs corregidos
print("\nüîÅ Regenerando archivo COCO JSON con IDs consistentes...")

# Debes tener esta lista cargada de antes (anotaciones_val)
# Si no la tienes en esta celda, c√°rgala desde donde la generaste
# anotaciones_val = [...]

json_coco_val = {
    "images": [],
    "annotations": [],
    "categories": [
        {"id": 1, "name": "aw"},
        {"id": 2, "name": "m"},
        {"id": 3, "name": "p"},
        {"id": 4, "name": "av"},
        {"id": 5, "name": "y"}
    ]
}

import cv2
ann_id_counter = 1
for nombre, img_id in img_id_map.items():
    ruta = VAL_IMG_DIR / nombre
    if not ruta.exists():
        print(f"‚ùå Imagen no encontrada: {nombre}")
        continue

    img = cv2.imread(str(ruta))
    if img is None:
        print(f"‚ö†Ô∏è No se pudo leer imagen: {nombre}")
        continue

    height, width = img.shape[:2]
    json_coco_val["images"].append({
        "id": img_id,
        "file_name": nombre,
        "width": width,
        "height": height
    })

    for ann in [a for a in anotaciones_val if a["nombre"] == nombre]:
        segmentacion = ann["segmentacion"]
        categoria_id = ann["categoria_id"]

        x_coords = segmentacion[0][0::2]
        y_coords = segmentacion[0][1::2]
        bbox = [
            float(min(x_coords)),
            float(min(y_coords)),
            float(max(x_coords) - min(x_coords)),
            float(max(y_coords) - min(y_coords))
        ]
        area = bbox[2] * bbox[3]

        json_coco_val["annotations"].append({
            "id": ann_id_counter,
            "image_id": img_id,
            "category_id": categoria_id,
            "segmentation": segmentacion,
            "bbox": bbox,
            "area": area,
            "iscrowd": 0
        })
        ann_id_counter += 1

# Guardar el JSON actualizado
with open(COCO_JSON_PATH, "w") as f:
    json.dump(json_coco_val, f, indent=2)

print(f"\n‚úÖ Nuevo archivo COCO JSON guardado en: {COCO_JSON_PATH}")
```
